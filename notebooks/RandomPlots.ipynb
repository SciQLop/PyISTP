```python
import os

os.environ["PYISTP_CDFLIB"] = 'pycdfpp'
#os.environ["PYISTP_CDFLIB"]='spacepy'
import pyistp
import numpy as np
import requests
#%matplotlib qt
%matplotlib widget
import matplotlib.pyplot as plt


def plot_timeserie(istp, var_names):
    plt.figure()
    if type(var_names) not in (list, tuple):
        var_names = (var_names,)
    for var_name in var_names:
        var = istp.data_variable(var_name)
        plt.plot(var.axes[0].values, var.values)
```

```python
thc_l2_efi = pyistp.load(requests.get(
    "https://spdf.gsfc.nasa.gov/pub/data/themis/thc/l2/efi/2020/thc_l2_efi_20200101_v01.cdf").content)
plot_timeserie(thc_l2_efi, 'thc_eff_dot0_gsm')
```

```python
psp_fld_l3_merged_scam_wf = pyistp.load(
    requests.get(
        "https://spdf.gsfc.nasa.gov/pub/data/psp/fields/l3/merged_scam_wf/2020/psp_fld_l3_merged_scam_wf_2020021006_v02.cdf").content)
plot_timeserie(psp_fld_l3_merged_scam_wf, 'psp_fld_l3_merged_scam_wf_uvw')
```

```python
psp_fld_l2_mag_sc_4_sa_per_cyc = pyistp.load(
    requests.get(
        "https://spdf.gsfc.nasa.gov/pub/data/psp/fields/l2/mag_sc_4_per_cycle/2021/psp_fld_l2_mag_sc_4_sa_per_cyc_20210111_v02.cdf").content)
plot_timeserie(psp_fld_l2_mag_sc_4_sa_per_cyc, 'psp_fld_l2_mag_SC_4_Sa_per_Cyc')
```

```python
voyager2_10s_mag = pyistp.load(
    requests.get(
        "https://spdf.gsfc.nasa.gov/pub/data/voyager/voyager2/magnetic_fields_cdaweb/mag_10s/1990/voyager2_10s_mag_19900824_v01.cdf").content)
plot_timeserie(voyager2_10s_mag, ('B1', 'B2', 'B3'))
plt.ylim(-5e-1, 5e-1)
```

```python
ge_1min_mag_plasma_sw_only = pyistp.load(
    requests.get(
        "https://spdf.gsfc.nasa.gov/pub/data/geotail/merged/magplasma_sw_1min_cdaweb/2005/ge_1min_mag_plasma_sw_only_20051221_v01.cdf").content)
plot_timeserie(ge_1min_mag_plasma_sw_only, ['VZ_GSE','VX_GSE','VY_GSE', 'V'])
plt.ylim(-1000,1000)
plot_timeserie(ge_1min_mag_plasma_sw_only, ['BY_GSM','BZ_GSM','BX_GSE'])
plt.ylim(-10,10)
plot_timeserie(ge_1min_mag_plasma_sw_only, ['SIGMA_By','SIGMA_ABS_B','SIGMA_B','SIGMA_Bx'])
plt.ylim(-1,3)
```

```python
ac_h2_sis = pyistp.load(
    requests.get(
        "https://cdaweb.gsfc.nasa.gov/pub/data/ace/sis/level_2_cdaweb/sis_h2/2021/ac_h2_sis_20210106_v09.cdf").content)

plot_timeserie(ac_h2_sis,
               ['cnt_Ni', 'cnt_Ne', 'cnt_Ar', 'cnt_C', 'cnt_Ca', 'cnt_Fe', 'cnt_He', 'cnt_N', 'cnt_O', 'cnt_Na',
                'cnt_Mg', 'cnt_Si', 'cnt_S'])
```
